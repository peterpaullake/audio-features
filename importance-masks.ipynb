# Learning a mask for an entire audio clip ("experience proves this")

```python
daf.make_audiofile_figure('meeting-3-audio/experience.wav',
                          'meeting-3-audio/experience-sentence-filtered.wav',
                          np.load('experience-sentence-mask.npy'))
```

```python
print('MASKED MFCCS: "%s"' % daf.transcribe_mfccs(np.load('experience-sentence-mask.npy') * \
                                                  daf.audiofile_to_mfccs('meeting-3-audio/experience.wav')))
titles = ['ORIGINAL AUDIO', 'FILTERED AUDIO']
filenames = ['meeting-3-audio/experience.wav', 'meeting-3-audio/experience-sentence-filtered.wav']
for title, filename in zip(titles, filenames):
    display(Audio(filename))
    print(title + ': "' + daf.transcribe_audiofile(filename) + '"')
```
